# Evaluation Overview

Make overview table of evaluation scores of all models, i.e., from the excel files from evaluate_model.py

```python
import os
import pandas as pd
```

```python
model_dir = os.path.join('..', 'models')
```

```python
scores = {}
df_scores = {}

for subdir, dirs, files in os.walk(model_dir):
    if 'evaluation_dev_centertiles' in os.path.basename(subdir):
        try:
            dff = pd.read_excel(os.path.join(subdir, "model_scores.xlsx"), index_col=0)#.loc[['mean', 'std'],:]
            dff.loc['median', :] = dff.loc[:4,:].median(axis=0)
            dff.loc['iqr', :] = dff.loc[:4,:].quantile(.75)-dff.loc[:4,:].quantile(.25)
            scores[os.path.basename(os.path.dirname(subdir))] = dff
            scores = {k.replace('alittleaug', 'littleaug') if 'alittleaug' in k else k: v for k, v in scores.items()}
        except:
            print(f"No evaluation file available for model {os.path.basename(os.path.dirname(subdir))}")
```

```python
df_scores = pd.concat(scores, axis=0).sort_index()
df_scores.to_excel(os.path.join('..', 'evaluation', 'scores_overview.xlsx'))
df_scores
```
